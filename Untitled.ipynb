```python
from  selenium import webdriver 
from bs4 import BeautifulSoup as bs
import re 
from translate import Translator
import string
import re
from time import  sleep
from selenium.common.exceptions import TimeoutException
import pickle
from collections import deque
from selenium.webdriver.common.by import By
import pandas as pd
import sys
sys.setrecursionlimit(10**6)
```

```python
#methods
def getfunc(link):
    while True:
        try:
            driver.get(link)
            x=driver.page_source
            return x
        except:
            print("time_out")
            sleep(60)
##################################################################
def ads_list(page_Source):
    try:
        soup=bs(page_Source,'html5lib')
        return[i['href'][0:23]+"en/"+i['href'][23:] for i in soup.find_all('a',class_="ads__item__ad--title")]

    except  Exception as e:
        print( str(e),"ad_list")

def next_page(page_source):
    try:
        x=bs(page_source,'html5lib').find_all('a',class_=re.compile("pageNextPrev"))
        if len(x)==1:
                if x[0].text.strip() =='Next Â»':
                    return x[0]['href']
        elif len(x)==2:
            return x[1]['href']
        else:
            return False
    except Exception as e :
        print(str(e)+' next')
        
def extract_data(page_source,i):
    try:
        res={}
        AddSoup=bs(page_source,'html5lib')
        date=AddSoup.find('span',class_="pdingleft10 brlefte5")
        Fdata=date.get_text()
        Fdata=Fdata.split(',')[1]
        ############################################
        price=AddSoup.find('strong',class_=re.compile("xxxx-large margintop7"))
        if soup.find('small',class_="block lheight15") != None:
            sign=1
        else:
            sign=0
        Fprice=price.get_text()
        ###################################################
        location=AddSoup.find('strong',class_="c2b small")
        Flocation=location.get_text().strip()
        ###################################################
        table=AddSoup.find(class_="details fixed marginbott20 margintop5 full")
        FiledsNames=[i.find('th') for i in table.find_all('td') if i!=None]
        FiledsNames=[i.text for i in FiledsNames if i!=None]
        FiledsValues=[i.text.split() for i in table.find_all('td',class_='value')]
        Vales=dict(zip(FiledsNames,FiledsValues))
        ###################################################
        title=AddSoup.find('h1',class_=re.compile("brkword lheight28"))
        Ftitle=title.text.strip()
        ##################################################
        viwers=AddSoup.find_all('div',class_="pdingtop10")
        Fviwers=viwers[1].find('strong').get_text()
        res={'price':Fprice,
            'location':Flocation,
            'data':Fdata,
            'title':Ftitle,
            'link':i,
            'viwers':Fviwers,
            'negotiable'  :sign             }
        res.update(Vales)
        return  res
    except  Exception as e:
        print("there is an error in vlaues",e)
        AddSoup=bs(page_source,'html5lib')
        date=AddSoup.find('span',class_="pdingleft10 brlefte5")
        ############################################
        price=AddSoup.find('strong',class_=re.compile("xxxx-large margintop7 block"))
        ###################################################
        location=AddSoup.find('strong',class_="c2b small")
        ###################################################
        table=AddSoup.find(class_="details fixed marginbott20 margintop5 full")
        ###################################################
        title=AddSoup.find('h1',class_=re.compile("brkword lheight28"))
        ##################################################
        viwers=AddSoup.find_all('div',class_="pdingtop10")
        res={'price':price,
            'location':location,
            'data':date,
            'title':title,
            'link':i,
            'vlaues':table,
             'viwers':viwers   }
        return  res
```

```python
chrome_options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
chrome_options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome('D:\coding-area\olx_scrapping\chromedriver.exe',chrome_options=chrome_options)

link="https://www.olx.com.eg/en/properties/"

try:
    with open("state.pkl",'rb') as file:
        all_data, stack, last_link = pickle.load(file)
except:
    all_data, stack, last_link = [], deque(), link

stack.appendleft(last_link)

while stack:
    link=stack.pop()
    last_link=link
    src=getfunc(link)
    print(len(all_data),link)
    if '/ad/' in link and link not in {x['link'] for x in all_data}:
        data=extract_data(src,link)
        all_data.append(data)
        with open("state.pkl",'wb') as file:
            state = all_data, stack, last_link
            pickle.dump(state, file)
        sleep(3)
    else:
        link=next_page(src)
        if link:
            stack.appendleft(link)
        stack.extendleft(ads_list(src))
```
